```python
import numpy as np
import pandas as pd
import os

pd.set_option('display.max_rows',81)
pd.set_option('display.max_columns',None)

datapath = "house-prices-advanced-regression-techniques"
trainfile = "train.csv"
testfile = "test.csv"

source = pd.read_csv(os.path.join(datapath,trainfile))
test_data = pd.read_csv(os.path.join(datapath,testfile))
train_data = source.loc[:,:'SalePrice']
target = source.loc[:,'SalePrice']
```

```python
data01 = pd.concat([train_data,test_data],axis=0)
data01 = data01.drop(columns=['Id','SalePrice'])
```

```python
data01
```

```python
data02 = data01
data02['MSSubClass'] = data02['MSSubClass'].astype(str)
```

```python
data03 = data02
numerical_columns = data03.select_dtypes(np.number).columns
numerical_columns
for i in numerical_columns :
    data03[i] = data03[i].fillna(data03[i].mean())
data03.loc[:,numerical_columns].isna().sum()
```

```python
data4 = data03
coloums_to_skip = ['Alley','BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2','FireplaceQu','GarageType','GarageFinish','GarageQual','GarageCond','PoolQC','Fence','MiscFeature','MasVnrType']
for i in coloums_to_skip :
    data4[i] = data4[i].fillna("None")
categorical_columns = data4.select_dtypes('object').columns
print(categorical_columns)
for i in categorical_columns :
    data4[i] = data4[i].fillna(data4[i].mode()[0])
data4.isna().sum()
```
