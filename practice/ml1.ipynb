```python
from sklearn import datasets
import numpy as np
from sklearn.preprocessing import  StandardScaler 
```

```python
iris = datasets.load_iris()
print(iris.feature_names)
X = iris.data[:,[2,3]]
y = iris.target
sc = StandardScaler()
sc.fit(X)
X = sc.transform(X)
print(np.unique(y))
```

```python
from matplotlib import pyplot as plt
x0 = X[y == 0][:,0]
y0 = X[y == 0][:,1]
x1 = X[y == 1][:,0]
y1 = X[y == 1][:,1]
x2 = X[y == 2][:,0]
y2 = X[y == 2][:,1]
plt.scatter(x0,y0,c = 'green',marker = 'o',label = 'Type 0(green circles)')
plt.scatter(x1,y1,c = 'red',marker = 's',label = 'Type 1(red squares)')
plt.scatter(x2,y2,c = 'black',marker = 'v',label = 'Type 2(black triangles)')
plt.legend()
plt.show()
```
